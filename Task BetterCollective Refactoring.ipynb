```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Specify the directory containing the CSV files
data_directory = r'C:\Users\Danica Filipovic\Desktop'

# Get a list of all CSV files in the directory
csv_files = [file for file in os.listdir(data_directory) if file.endswith('.csv')]

# Loop through each CSV file in the directory
for csv_file in csv_files:
    # Construct the full path to the CSV file
    file_path = os.path.join(data_directory, csv_file)

    # Read the data from the CSV file into a Pandas DataFrame
    df = pd.read_csv(file_path)

    # Extract the season from the CSV file name
    # (Assuming the file name contains the season information, adjust this based on your file naming convention)
    season = csv_file.split('_')[0]
    
    # Calculate the average age per team
    average_age_per_team_season = df.groupby(['Tm'])['Age'].mean()

    # Visualize the average age per team using a bar chart
    average_age_per_team_season.plot(kind='bar', title=f'Average Age per Team - {season}')
    plt.ylabel('Average Age')
    plt.xlabel('Team')
    plt.show()

    # Print the result
    print(f"\nAverage age per team for {season}:")
    print(average_age_per_team_season)
    
    # Calculate the average and maximum age per season
    average_max_age_per_season = df['Age'].agg(['mean', 'max'])

    # Visualize the average and maximum age per season using a bar chart
    average_max_age_per_season.plot(kind='bar', title=f'Average and Maximum Age per Season - {season}')
    plt.ylabel('Age')
    plt.xlabel('Statistic')
    plt.show()

    # Print the result
    print(f"\nAverage and maximum age per season for {season}:")
    print(average_max_age_per_season)
```
